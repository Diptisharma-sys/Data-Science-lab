```python
## file.txt ma five ko multiplication

def multiplication_table():
    table = ""
    for i in range(1,11):
        n = (4*(i-1 ))
        spaces = " " * n
        table += f"\n {spaces} 5 x {i} = {5*i}\n "
    return table

file_name = "file.txt"
with open('file.txt',"w") as f:
    f.write(multiplication_table())

with open('file.txt','r') as f:
    content = f.read()
    print(content)
    f.close()

        
```
