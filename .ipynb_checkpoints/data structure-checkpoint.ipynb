```python
# Stack using library

from array import array
stack = array('i',[])

stack.append(100)
stack.append(200)
stack.append(300)
stack.append(400)

print("Stack",stack)

removed = stack.pop()
print("Removed:",removed)

print(stack[-1])


```

```python
# STACK
stack = []

def push(item):
    global stack
    stack = [item] + stack   
    print(item, "pushed")

def pop():
    global stack
    if len(stack) == 0:
        print("Stack is empty")
    else:
        removed = stack[0]   
        stack = stack[1:]   
        print("Popped:", removed)

def display():
    print("Stack:", stack)


display()       
push(20)
push(30)
push(40)
display()       
pop()
display()      
```
