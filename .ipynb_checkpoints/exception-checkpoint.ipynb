## Exceptions

### types
### FileNotFound
### ZeroDivisionError
### ValueError
### PermissionError

```python
try:
    x=10/0
except ZeroDivisionError:
    print("can't divisible by zero")
```

```python
try:
    x = 10 / 0
except Exception as er:
    print(er)
```

```python
try:
    f = open("data.txt", "r")
    data = f.read()
    f.close()
except FileNotFoundError as e:
    print("File not found")
    print(e)
else:
    print("File content:")
    print(data)
```

```python
try:
    f=open("unknown.txt")
except FileNotFoundError:
    print("No such file")
except PermissionError:
    print("Permission denied")
```

```python
### d is integer
import re
text = "My phone number is 9841675545"
result = re.search(r"\d",text)
print(result)
```
