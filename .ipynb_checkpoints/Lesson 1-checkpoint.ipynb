# Describe all these operators and example of each with output

## arithmetic operators

```python
a = 10
b = 3
print(a + b)   # Addition
print(a - b)   # Subtraction
print(a * b)   # Multiplication
print(a / b)   # Division
print(a % b)   # Modulus
print(a ** b)  # Exponentiation
print(a // b)  # Floor Division
```

## comparison operators

```python
a = 5
b = 8
print(a == b)
print(a != b)
print(a > b)
print(a < b)
print(a >= b)
print(a <= b)
```

## logical Operators

```python
x = 5
y = 10
print(x > 3 and y > 5)
print(x > 7 or y > 5)
print(not(x > 3 and y > 5))
```

## assignment operators

```python
x = 10 

x += 5
print(x)
x -= 3
print(x)
x *= 2
print(x)
x /= 4
print(x)
x %= 3
print(x)
```

## bitwise operators

```python
a = 5   # 0101
b = 3   # 0011
print(a & b)   # AND
print(a | b)   # OR
print(a ^ b)   # XOR
print(~a)      # NOT
print(a << 1)  # Left shift
print(a >> 1)  # Right shift
```

## membership operators

```python
fruits = ["apple", "banana", "mango"]
print("apple" in fruits)
print("grape" not in fruits)
```

## identity operators

```python
x = [1, 2, 3]
y = [1, 2, 3]
z = x

print(x is y)
print(x is z)
print(x is not y)
```

## Functions

```python
def add(a,b):
    return a+b
def sub(a,b):
    return a-b

print(add(2,3))
print(sub(3,1))
```

```python
def is_prime(n):
    if n<=1: return False
    for i in range(2,n):
        if n%i ==0 :
            return False
    return True
print(is_prime(8))
```

```python
def function(n):
    if n == 0:
        return
    else:
        function(n-1)
        print(n)
function(10)
```

 # write a program that asks for three numbers abd prints largest using a function

```python
def function (a,b,c):
    if a>=b and a>=c:
        return a
    elif b>=a and b>=c:
        return b
    else:
        return c 

a = int(input("enter your first number"))
b = int(input("enter your second number"))
c = int(input("enter your third number"))

largest = function(a,b,c)
print( "largest :",largest)
```

```python
def largest(arr_):
    max_number = arr_[0]
    for a in arr_:
        if a>max_number:
            max_number = a
        return max_number

largest([1,2,3,4,5])
```
