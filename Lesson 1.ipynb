 # write a program that asks for three numbers abd prints largest using a function

```python
def function (a,b,c):
    if a>=b and a>=c:
        return a
    elif b>=a and b>=c:
        return b
    else:
        return c 

a = int(input("enter your first number"))
b = int(input("enter your second number"))
c = int(input("enter your third number"))

largest = function(a,b,c)
print( "largest :",largest)
```

```python
def largest(arr):
    max_number = arr[0]
    for a in arr:
        if a > max_number:
            max_number = a
    return max_number

largest([1,2,3,4,5])
```

```python
def largest(arr):
    max_number = arr[0]
    for a in arr:
        if a > max_number:
            max_number = a
    return max_number

# Taking input from user
numbers = input("Enter numbers separated by space: ").split()  # input as string and split into list
numbers = [int(num) for num in numbers]  # convert each element to integer

# Find and print largest number
print("Largest number is:", largest(numbers))
```

# Roll a dice 10 times using random randint() and count how many times you get 6

```python
import random

count_6 = 0

for i in range(10):  # i goes from 0 to 9 (10 rolls)
    roll = random.randint(1, 6)  # dice roll between 1 and 6
    print(f"Roll {i+1}: {roll}")  # shows which roll it is
    if roll == 6:
        count_6 += 1

print("Number of times 6 appeared:", count_6)

```

# Ask the user to input a word and count vowels and consonants using a function

```python
def checkword(word):
    vowels = 0
    consonants = 0

    for letter in word.lower():
        if letter in "aeiou":
            vowels += 1
        elif letter.isalpha():
            consonants += 1

    print( "vowels:", vowels)
    print( "consonants:", consonants)


word = input("enter a word")

checkword(word)
```

# Build a simple grade calculator for multiple students. Input their marks in a list and print the grade using a function

```python
def checkGrade(marks):
    if marks >= 90:
        print("Grade: A+")
    elif marks >= 80:
        print("Grade: A")
    elif marks >= 70:
        print("Grade: B")
    elif marks >= 60:
        print("Grade: C")
    elif marks >= 50:
        print("Grade: D")
    else:
        print("Grade: F")

# Example usage
score = int(input("Enter your marks: "))
checkGrade(score)
```

```python
# Function to calculate grade based on marks
def calculate_grade(marks):
    if marks >= 90:
        return 'A+'
    elif marks >= 80:
        return 'A'
    elif marks >= 70:
        return 'B'
    elif marks >= 60:
        return 'C'
    elif marks >= 50:
        return 'D'
    else:
        return 'F'

# Input marks from user
marks_list = input("Enter marks of students separated by space: ").split()
marks_list = [int(m) for m in marks_list]

# Print grades for each stude
for i in range(len(marks_list)):
    grade = calculate_grade(marks_list[i])
    print("Student", i+1, "Marks =", marks_list[i], "Grade =", grade)

```
