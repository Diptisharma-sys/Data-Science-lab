```python
dict ={
    "nepali song":"sathi ma timro",
    "english song":"ladies and gentle man",
    "hindi song":"bin tere"
    }

for key,value in dict.items():
    print(f"{key}:{value}")
```

```python
dict ={
    "nepali song":"sathi ma timro",
    "english song":"ladies and gentle man",
    "hindi song":"bin tere",
    "dict2" :{
    "another song":"Lover"
    }
}

for key, value in dict.items():
    if type(value) == dict:  # check if it is a nested dictionary
        for subkey, subvalue in value.items():  # go inside it
            print(subkey + " : " + subvalue)
    else:
        print(f"{key}:{value}")
```

```python
songs_dict = {
    "nepali song": "sathi ma timro",
    "english song": "ladies and gentle man",
    "hindi song": "bin tere",
    "dict2": {
        "another song": "Lover"
    }
}

for key, value in songs_dict.items():
    if isinstance(value, dict):  # check if value is a dictionary
        for subkey, subvalue in value.items():
            print(f"{subkey} : {subvalue}")
    else:
        print(f"{key} : {value}")
```

```python
s = "Python"
stack = list(s)       # Convert string to list (stack)
reversed_s = ""

while stack:          # While stack is not empty
    reversed_s += stack.pop()  # Pop from stack and append to reversed_s

print(reversed_s)     # Output: nohtyP
```
