### GCD

```python


def gcd(a,b):
    factors_a =[]
    factors_b=[]
    
    for i in range( 1,a+1):
        if a%i ==0 :
            factors_a.append(i)
    
    for i in range(1,b+1):
        if b%i==0:
            factors_b.append(i)

    print("factors of ",a ,":",factors_a)
    print("factors of ",b ,":",factors_b)

    common_factors = []
    for i in factors_a:
        if i in factors_b:
            common_factors.append(i)

    print("common factors:",common_factors)
    return max(common_factors)
    
a= int(input("enter a number:"))
b=int(input("enter another number:"))

GCD = gcd(a,b)
print("GCD is :",GCD)


    
        
    
```

### GCD EFFICIENTLY

```python
def gcd(a, b):
    smallest = min(a, b)
    for i in range(smallest, 0, -1):
        if a % i == 0 and b % i == 0:
            return i  

a = int(input("Enter first number: "))
b = int(input("Enter second number: "))

print("GCD is:", gcd(a, b))
```

### Write a program that checks if a year is a leap year. 

```python
year = int(input("Enter a year: "))

if (year % 400 == 0) or (year % 4 == 0 and year % 100 != 0):
    print(year, "is a leap year")
else:
    print(year, "is not a leap year")
```

###	Write a program that takes a temperature in Celsius, and converts it to Fahrenheit and Kelvin, based on the choice of user. 

```python
celsius = float(input("Enter temperature in Celsius: "))

print("Choose the conversion:")
print("1. Fahrenheit")
print("2. Kelvin")

choice = int(input("Enter your choice (1 or 2): "))

if choice == 1:
    fahrenheit = (celsius * 9/5) + 32
    print("Temperature in Fahrenheit:", fahrenheit)
elif choice == 2:
    kelvin = celsius + 273.15
    print("Temperature in Kelvin:", kelvin)
else:
    print("Invalid choice!")
```
