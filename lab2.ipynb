### Ask the user to input a list of movies with ratings like ("Titanic",8),("Inception",9),....
### Compute the average rating, find the highest-rated movie, and list all movies with rating above the average

```python
def movie_analyzer():
    movies = []

    print("Enter movies with ratings like: Titanic,8")
    print("Type 'finished' when done.\n")

    while True:
        user_input = input("Enter movie with rating: ")

        if user_input.lower() == "finished":
            break

        try:
            name, rating = user_input.split(",")
            name = name.strip()
            rating = float(rating.strip())
        except:
            print("Invalid format! Use Movie,Rating (Example: Titanic,8)")
            continue
        
        movies.append((name, rating))

    if not movies:
        print("No movies entered.")
        return

    # Average rating
    ratings = [r for _, r in movies]
    avg_rating = sum(ratings) / len(ratings)

    # Highest-rated movie
    highest_movie = max(movies, key=lambda x: x[1])

    # Movies above average
    above_avg = [m for m in movies if m[1] > avg_rating]

    print("RESULTS")
    print(f"Average Rating: {avg_rating:.2f}")
    print(f"Highest Rated Movie: {highest_movie[0]} ({highest_movie[1]})")

    print("Movies Above Average Rating:")
    if above_avg:
        for m, r in above_avg:
            print(f"- {m} ({r})")
    else:
        print("No movies above average.")


movie_analyzer()
```

### Unique Words Collector Take a paragraph input from the user. Split it into words, remove duplicates, sort them alphabetically, and count the total number of unique words.

### Student Marks Manager
### Store student names as keys and marks (list of integers) as values in a dictionary.Compute each studentâ€™s average and grade (A/B/C/D).Print the top 2 students based on average marks.

```python
students = {
    "Pragati": [90, 80, 90, 70, 60],
    "Indu": [80, 60, 50, 90],
    "Prajina": [90, 50, 70, 80],
    "Dipti": [80, 60, 50, 40]
}

# Function to determine grade
def get_grade(avg):
    if avg >= 80:
        return 'A'
    elif avg >= 70:
        return 'B'
    elif avg >= 60:
        return 'C'
    else:
        return 'D'

student_results = {}

for student, marks in students.items():
    average = sum(marks) / len(marks)
    grade = get_grade(average)
    student_results[student] = {'average': average, 'grade': grade}

# Print each student's average and grade
for student, result in student_results.items():
    print(f"{student}: Average = {result['average']:.2f}, Grade = {result['grade']}")

# Top 2 students
top_students = sorted(student_results.items(), key=lambda x: x[1]['average'], reverse=True)[:2]

print("\nTop 2 students based on average marks:")
for student, result in top_students:
    print(f"{student}: Average = {result['average']:.2f}, Grade = {result['grade']}")
```

### 4. Inventory Tracker Imagine a small store inventory like {"apple": 10, "banana": 5, "milk": 2}.Program should allow adding new items, selling items (subtract quantity), and printing items that are low in stock (<3).

```python
inventory ={
    "apple":10,
    "banana":5,
    "milk":2
}
def addnewitems():
    while True:
        items = input("Enter the item you want to add (type 'done' to stop): ")

        if items.lower() == "done":
            break

        qty = int(input(f"Enter quantity of {items}: "))

        if items in inventory:
            inventory[items] += qty
        else:
            inventory[items] = qty

        print(f"{qty} of {items} added successfully.")


def sellitems():
    while True:
        items = input("Enter the item you want to sell (type 'done' to stop): ")

        if items.lower() == "done":
            break

        qty = int(input(f"Enter quantity of {items}: "))

        if items not in inventory:
            print("Item not available!")
        elif inventory[items] < qty:
            print(f"Not enough stock! Available: {inventory[items]}")
        else:
            inventory[items] -= qty
            print(f"{qty} of {items} sold successfully!")


def lowstock():
    print("Items That are low in stock")
    for item, qty in inventory.items():
        if qty < 3:
            print(f"{item}: {qty}")
        
def main():
    print("1 .add new items")
    print("2 .sell items")
    print("3 .view low stock items")
    option = int(input("enter the option you want to perform"))

    if option==1:
        addnewitems()
    elif option==2:
        sellitems()
    elif option==3:
        lowstock()
    else:
        print("invalid ")

main()


    




```

### Word Reverser GameAsk the user for a list of words. Reverse each word only if its length is even.Print the new list of words after processing.

```python
words = input("Enter words separated by comma (,): ").split(",")

# remove spaces
words = [w.strip() for w in words]
print(words)

result = []

for w in words:
    if len(w) % 2 == 0:
        result.append(w[::-1])   
    else:
        result.append(w)

print(result)
```

### Simple Path ValidatorRepresent a small map as a dictionary like {"A": {"B", "C"}, "B": {"A", "D"}, "C": {"A", "D"}, "D": {"B", "C"}}.Ask the user to input a path, e.g., "A","C","D".Check if each consecutive step is connected and print "Valid path" or "Invalid path"

```python
# Represent the map as a dictionary
map_graph = {
    "A": {"B", "C"},
    "B": {"A", "D"},
    "C": {"A", "D"},
    "D": {"B", "C"}
}

# Ask the user to input a path
user_input = input("Enter a path as a list of nodes (e.g., A C D): ")
path = user_input.split()  # Convert input string into a list of nodes

# Initialize validity flag
is_valid = True

# Check consecutive steps in the path
for i in range(len(path) - 1):
    current_node = path[i]
    next_node = path[i + 1]
    # If next_node is not connected to current_node
    if next_node not in map_graph.get(current_node, {}):
        is_valid = False
        break

# Print result
if is_valid:
    print("Valid path")
else:
    print("Invalid path")
```
